# Seaborn Visualization
* Seaborn is used for advanced graphs and visualization
* A complement to matplotlib
* import **seaborn** as **sns**  

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
%matplotlib inline
sns.set()
```

```python
new_iris = sns.load_dataset('iris')
```

```python
new_iris.head()
```

```python
new_iris.head(3)
```

```python
new_iris.petal_length.plot()
```

```python
new_iris.plot()
```

```python
new_d = pd.Series(np.random.rand(7), index=list('ABCDEFG'))
new_d
```

```python
new_d.plot.bar(color='r', alpha=1)
```

```python
new_d.plot.bar(color='g', alpha=0.5)
```

```python
new_d.plot.bar(color='b', alpha=0.1)
```

```python
new_d.plot.barh(color='y', alpha=0.7)
```

```python
new_loading = sns.load_dataset('tips')
new_loading.head(4)
```

```python
size_of_day = pd.crosstab(new_loading['day'], new_loading['size'])
size_of_day
```

```python
size_of_day.plot.barh()
```

```python
size_of_day.plot.hist()
```

```python
size_of_day.plot.bar()
```

# Seaborn statistical graph
* Using Seaborn  to create heatmaps
* The heatmap is used to provide a colored visual summary of data
* Use **sb.heatmap(new_data)** to create heatmaps

```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
new_data = np.random.rand(6, 8)
 
ht_map = sb.heatmap(new_data)
plt.show()
```
