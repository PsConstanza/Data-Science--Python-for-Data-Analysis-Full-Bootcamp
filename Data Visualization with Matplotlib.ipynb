# Matplotlib Data Visualization
* Using matplotlib library to visualize data with post and graphs... It's an open source library
* import **matplotlib.pyplot** as plt
* import **numpy** as np

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
new_data = np.arange(9)
new_data
```

```python
#%matplotlib notebook
#%matplotlib inline
```

```python
plt.plot(new_data)
```

```python
plt.title('The new data grap')
```

```python
plt.xlabel('x label name')
```

```python
plt.ylabel('y label name')
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 4, 9, 16, 25, 36, 49]
```

```python
plt.plot(x, y)
```

```python
data1 = np.arange(2, 7, 0.3)
plt.plot(data1, data1, 'r--', data1, data1**2, 'bs', data1, data1**3, 'g^')
```

```python
plt.savefig("newvisualize.png")
```

# Histograms
* Using matplotlib.**pyplot** to show frequency distributions
* plt.**hist()**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
new_h = np.random.normal(150, 15, 230)
new_h
```

```python
plt.hist(new_h)
plt.show()
```

# Pie Charts
* Using matplotlib.**pyplot** to create Pie Charts
* **plt.pie**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([37, 27, 27, 17])
ch_labels = ['first', 'second', 'third', 'fourth']
```

```python
plt.pie(x, labels = ch_labels)
plt.show()
```
